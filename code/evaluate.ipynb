## Setup

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the plot style for better visuals
sns.set(style="whitegrid")

# Path to the CSV directory
csv_path = '../data/csv/'

# List of CSV files to analyze
csv_files = [
    'movie_table_stats.csv',
    'top_rated_movies.csv',
    'most_popular_movies.csv',
    'top_directors_by_avg_rating.csv',
    'genre_popularity.csv',
    'top_actors_by_rating.csv',
    'yearly_movie_trend.csv',
    'adult_vs_non_adult_trend.csv',
    'top_movie_per_year.csv',
    'genre_trend_over_time.csv',
    'most_consistent_directors.csv',
    'breakout_actors.csv',
    'highest_grossing_genres.csv',
    'director_vs_writer_success.csv',
    'underrated_movies.csv'
]

# Function to read all CSVs into a dictionary of DataFrames
def load_csvs(files):
    return {file: pd.read_csv(csv_path + file) for file in files}

# Load the CSVs
dataframes = load_csvs(csv_files)
```

## Evaluation

## Visual Analysis

### Top Rated Movies

```python
# Visualize the top-rated movies
top_rated_movies_df = dataframes['top_rated_movies.csv']
plt.figure(figsize=(10, 6))
sns.barplot(x='rating', y='movie_title', data=top_rated_movies_df.head(10))
plt.title("Top 10 Rated Movies")
plt.xlabel("Rating")
plt.ylabel("Movie Title")
plt.show()
```

### Most Popular Movies

```python
# Visualize the most popular movies
most_popular_movies_df = dataframes['most_popular_movies.csv']
plt.figure(figsize=(10, 6))
sns.barplot(x='popularity_score', y='movie_title', data=most_popular_movies_df.head(10))
plt.title("Top 10 Most Popular Movies")
plt.xlabel("Popularity Score")
plt.ylabel("Movie Title")
plt.show()
```

### Genre Popularity

```python
# Genre popularity analysis
genre_popularity_df = dataframes['genre_popularity.csv']
plt.figure(figsize=(12, 8))
sns.barplot(x='movie_count', y='genre_name', data=genre_popularity_df)
plt.title("Genre Popularity by Movie Count")
plt.xlabel("Number of Movies")
plt.ylabel("Genre")
plt.show()
```

### Directors with the Highest Average Rating

```python
# Top directors by average rating
top_directors_df = dataframes['top_directors_by_avg_rating.csv']
plt.figure(figsize=(10, 6))
sns.barplot(x='avg_rating', y='director_name', data=top_directors_df.head(10))
plt.title("Top 10 Directors by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("Director Name")
plt.show()
```

### Breakout Actors

```python
# Analyze breakout actors
breakout_actors_df = dataframes['breakout_actors.csv']
plt.figure(figsize=(10, 6))
sns.barplot(x='rating_growth', y='actor_name', data=breakout_actors_df)
plt.title("Top Breakout Actors (Rating Growth)")
plt.xlabel("Rating Growth")
plt.ylabel("Actor Name")
plt.show()
```

## Trend Analysis

### Top Movie Per Year

```python
top_movie_df = dataframes['top_movie_per_year.csv']

# Line plot showing the trend of the highest-rated movies over the years
plt.figure(figsize=(14, 8))
sns.lineplot(x='year', y='rating', data=top_movie_df, marker='o', color='b')
plt.title('Trend of Highest-Rated Movies Over the Years')
plt.xlabel('Year')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Scatter plot to see the distribution of ratings across years
plt.figure(figsize=(14, 8))
sns.scatterplot(x='year', y='rating', data=top_movie_df, s=100, color='orange', edgecolor='black')
plt.title('Top Movie Ratings Distribution by Year')
plt.xlabel('Year')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Yearly Movie Release Trend

```python
# Yearly movie release trend
yearly_movie_trend_df = dataframes['yearly_movie_trend.csv']
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='total_movies', data=yearly_movie_trend_df, label="Total Movies")
plt.title("Yearly Movie Release Trend")
plt.xlabel("Year")
plt.ylabel("Movie Count")
plt.legend()
plt.show()
```

### Highest Grossing Genres

```python
grossing_genres_df = dataframes['highest_grossing_genres.csv']

# Plot 1: Bar plot for Total Popularity of Genres
plt.figure(figsize=(14, 8))
sns.barplot(x='total_popularity', y='genre_name', data=grossing_genres_df, hue='genre_name', palette='plasma', dodge=False)
plt.title('Total Popularity by Genre')
plt.xlabel('Total Popularity')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()

# Plot 2: Bar plot for Total Vote Count of Genres
plt.figure(figsize=(14, 8))
sns.barplot(x='total_votes', y='genre_name', data=grossing_genres_df, hue='genre_name', palette='coolwarm', dodge=False)
plt.title('Total Vote Count by Genre')
plt.xlabel('Total Vote Count')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()

# Plot 3: Scatter plot for Total Popularity vs Total Vote Count by Genre
plt.figure(figsize=(14, 8))
sns.scatterplot(x='total_popularity', y='total_votes', data=grossing_genres_df, hue='genre_name', palette='Set1', s=100)
plt.title('Genre Popularity vs. Vote Count')
plt.xlabel('Total Popularity')
plt.ylabel('Total Vote Count')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
